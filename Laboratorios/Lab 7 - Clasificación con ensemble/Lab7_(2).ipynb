4. Considerando:

*Verdadero positivo (TP):* Instancias positivas que fueron correctamente clasificadas.

*Verdadero negativo (TN):* Instancias negativas que fueron correctamente clasificadas.

*Falso positivo (FP):* Instancias negativas que fueron incorrectamente clasificadas como positivas.

*Falso negativo (FN):* Instancias positivas que fueron incorrectamente clasificadas negativas.

  * Accuracy: es la proporci칩n del conjunto de prueba que est치 clasificado correctamente y se calcula como:

$$
\displaystyle \small accuracy = \frac{TP + TN}{TP + FP + TN + FN}
$$

  * Precision: es una medida de la exactitud, o el porcentaje de instancias clasificadas como positivas que realmente lo son y se calcula como:

$$
\displaystyle \small precision = \frac{TP}{TP + FP}
$$

  * Recall: es la proporci칩n de instancias positivas que son clasificadas correctamente y se calcula como:
  
$$
\displaystyle \small recall = \frac{TP}{TP + FN}
$$

  * F1 score: es la combinaci칩n entre precision y recall en una solo medida. Esto otorga igual peso a la precisi칩n y la exhaustividad y se calcula como:

$$
\displaystyle \small F_1 = \frac{2\times precision \times recall}{precision + recall}
$$

**Nota**:  Renac칤n les recalca que **no deben utilizar los valores ```Sub``` y ```Res``` de esta columna**.

```python
ataque = ['ST', 'CF']
```

La elecci칩n del clasificador se justificar en base a la siguiente [gu칤a](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html) y se deben comentar los resultados obtenidos en la clasificaci칩n.

**Tareas:** [1 punto por tarea]

**Respuesta:**

````python
# Funci칩n para etiquetar las posiciones
def label_positions(position):
    if position in ['ST', 'CF']:
        return 'ataque'
    elif position in ['RW', 'CAM', 'LW']:
        return 'central_ataque'
    elif position in ['RM', 'CM', 'LM']:
        return 'central'
    elif position in ['RWB', 'CDM', 'LWB']:
        return 'central_defensa'
    elif position in ['RB', 'CB', 'LB']:
        return 'defensa'
    elif position == 'GK':
        return 'arquero'
    else:
        return None  # Excluimos las posiciones 'Sub' y 'Res'

df['label_pos'] = df['Club_Position'].apply(label_positions)

df = df[df['label_pos'].notna()]

class_counts = df['label_pos'].value_counts()
class_counts
````

Por lo de la gu칤a deber칤a ser SVC?

* Arboles de decisi칩n (dt):  es un modelo de aprendizaje supervisado que organiza los datos en una estructura de 치rbol, donde cada nodo corresponde a una caracter칤stica del conjunto de datos y cada rama representa una decisi칩n o regla basada en el valor de esa caracter칤stica. El objetivo es dividir los datos de manera que se maximice la ganancia de informaci칩n o se minimice la impureza (entrop칤a o Gini). Este proceso contin칰a hasta que se alcanza un criterio de parada, como un n칰mero m칤nimo de muestras en un nodo o una m치xima profundidad del 치rbol.

* Random Forest (rf): es un modelo que crea m칰ltiples 치rboles de decisi칩n en paralelo, cada 치rbol se entrena con un subconjunto aleatorio de los datos de entrenamiento, y en cada nodo, se selecciona aleatoriamente un subconjunto de caracter칤sticas para determinar la mejor divisi칩n. La predicci칩n final del Random Forest se obtiene mediante el promedio (para regresi칩n) o el voto mayoritario (para clasificaci칩n) de todos los 치rboles.

* Extra Tress (et): es un modelo similar a Random Forest, pero con una mayor aleatoriedad en el proceso de construcci칩n de los 치rboles. Al igual que en Random Forest, se construyen m칰ltiples 치rboles de decisi칩n en paralelo utilizando subconjuntos aleatorios de los datos de entrenamiento. Sin embargo, la diferencia clave es que en lugar de buscar la mejor divisi칩n en cada nodo, Extra Trees elige las divisiones de manera completamente aleatoria.

````python
from pycaret.classification import *
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score
import pandas as pd
import numpy as np

def create_detailed_selection_process(df, target='label'):
    # Obtener todas las caracter칤sticas originales
    all_features = [col for col in df.columns if col != target]
    ignored_features_initial = ['Name', 'Nationality', 'National_Position', 'label_pos']

    # Configuraci칩n inicial (con ajuste de clases desbalanceadas)
    clf = setup(data=df, target=target, session_id=7, fold_shuffle=True,
                ignore_features=ignored_features_initial, verbose=False, fix_imbalance=True)  # Ajuste de clases desbalanceadas

    # Entrenar modelo inicial
    xgb_model = create_model('xgboost')

    # Obtener importancias originales
    initial_features = get_config('X').columns.tolist()
    initial_importances = xgb_model.feature_importances_
    original_importance_dict = dict(zip(initial_features, initial_importances))

    # A침adir cero para caracter칤sticas ya ignoradas
    for feat in ignored_features_initial:
        original_importance_dict[feat] = 0

    # Preparar listas para almacenar resultados
    results = []
    ignored_features = ignored_features_initial.copy()
    accuracy_list = []
    recall_list = []
    f1_list = []
    precision_list = []

    # Funci칩n para evaluar el modelo
    def evaluate_model(predictions):
        try:
            y_true = predictions[target]
            y_pred = predictions['prediction_label']  # Usar 'Label' en lugar de 'prediction_label'
            acc = accuracy_score(y_true, y_pred)
            rec = recall_score(y_true, y_pred, average='weighted')
            f1 = f1_score(y_true, y_pred, average='weighted')
            prec = precision_score(y_true, y_pred, average='weighted')
            return acc, rec, f1, prec
        except Exception as e:
            print(f"Error en evaluaci칩n: {str(e)}")
            return 0.0, 0.0, 0.0, 0.0

    iteration = 0
    while True:
        current_features = get_config('X').columns.tolist()
        predictions = predict_model(xgb_model)
        accuracy, recall, f1, precision = evaluate_model(predictions)

        # Almacena las m칠tricas en las listas
        accuracy_list.append(accuracy)
        recall_list.append(recall)
        f1_list.append(f1)
        precision_list.append(precision)

        # Obtener importancias actuales
        current_importances = xgb_model.feature_importances_
        current_importance_dict = dict(zip(current_features, current_importances))

        # Crear diccionario para esta iteraci칩n
        iteration_dict = {
            'Model_Iteration': iteration,
            'Removed_Feature': 'None' if iteration == 0 else removed_feature,
            'Original_Importance': 0 if iteration == 0 else original_importance_dict[removed_feature],
            'Features_Used': len(current_features),
            'Features_Ignored': len(ignored_features),
            'Accuracy': accuracy,
            'Recall': recall,
            'F1_Score': f1,
            'Precision': precision
        }

        # A침adir importancias actuales para todas las caracter칤sticas originales
        for feature in all_features:
            importance_key = f'Importance_{feature}'
            if feature in current_importance_dict:
                iteration_dict[importance_key] = current_importance_dict[feature]
            else:
                iteration_dict[importance_key] = np.nan

        results.append(iteration_dict)

        if len(current_features) <= 1:
            break

        # Encontrar la pr칩xima caracter칤stica a eliminar (la de menor importancia original)
        available_features = set(current_features) - set(ignored_features)
        if not available_features:
            break

        removed_feature = min(available_features, key=lambda x: original_importance_dict.get(x, float('inf')))
        ignored_features.append(removed_feature)

        # Reentrenar modelo sin reiniciar el setup
        clf = setup(data=df, target=target, session_id=7, fold_shuffle=True,
                   ignore_features=ignored_features, verbose=False, fix_imbalance=True)  # Mantener ajuste para clases desbalanceadas
        xgb_model = create_model('xgboost')

        iteration += 1

    # Crear DataFrame final
    results_df = pd.DataFrame(results)

    # A침adir las listas de m칠tricas al DataFrame
    results_df['Accuracy'] = accuracy_list
    results_df['Recall'] = recall_list
    results_df['F1_Score'] = f1_list
    results_df['Precision'] = precision_list

    # Reordenar columnas
    importance_cols = [col for col in results_df.columns if col.startswith('Importance_')]
    base_cols = ['Model_Iteration', 'Removed_Feature', 'Original_Importance',
                 'Features_Used', 'Features_Ignored']
    final_cols = base_cols + importance_cols + ['Accuracy', 'Recall', 'F1_Score', 'Precision']

    results_df = results_df[final_cols]
    return results_df

# Ejecutar el proceso
detailed_results = create_detailed_selection_process(df)

# Guardar resultados
detailed_results.to_csv('detailed_feature_selection_results.csv', index=False)

# Mostrar primeras filas
print(detailed_results.head())

# Mostrar estad칤sticas b치sicas
print("\nEstad칤sticas de m칠tricas:")
print(detailed_results[['Accuracy', 'Recall', 'F1_Score', 'Precision']].describe())

# Encontrar el mejor modelo seg칰n F1-Score
best_model = detailed_results.loc[detailed_results['F1_Score'].idxmax()]
print("\nMejor modelo:")
print(f"Iteraci칩n: {best_model['Model_Iteration']}")
print(f"Caracter칤sticas usadas: {best_model['Features_Used']}")
print(f"F1-Score: {best_model['F1_Score']:.4f}")
````
